```python
# Import Data
df = pd.read_csv('IMDB Review.csv')
python_mask = np.array(PIL.Image.open('rth.png'))
```

```python
# Unify Text from all weeks
text = df['tweet_text'].str.cat(sep=' ')
```

```python
#set the stopwords list
stopwords= set(STOPWORDS)
new_words = []
new_stopwords=stopwords.union(new_words)
```

```python
# Size of Word Cloud
plt.rcParams["figure.figsize"] = (10,10)
```

```python
# Unify Text from tweets of different sentiment
text = df[df['Sentiment']=='positive']['tweet_text'].str.cat(sep=' ')
```

```python
wc = WordCloud(mask=python_mask, min_font_size=5, max_words=230, background_color="white", stopwords=new_stopwords, colormap='plasma').generate(text)
plt.plot()
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
